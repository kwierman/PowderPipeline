```python
pass_name = None
price = None
age = None
```

```python
delimiter = "**"
age_tokens = ["Adult", "Child", "Teen", "Military", "Veteran"]
for line in lines:
    tokens = line.split(delimiter)
    for token in tokens:
        if "Pass" in token:
            pass_name = token.strip()
        elif "$" in token:
            price = token.strip()
        else:
            for age_token in age_tokens:
                if age_token in token:
                    age = token.strip()
        if pass_name and price and age:
            print(f"Pass Name: {pass_name}, Price: {price}, Age Group: {age}")
            pass_name = None
            price = None
```
