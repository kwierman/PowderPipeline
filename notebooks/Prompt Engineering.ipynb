```python
import yaml
```

```python
api_key = yaml.safe_load(open('../config.yaml'))['openairouter_api_key']
openairouter_model_name= 'google/gemini-3-pro-preview'
openairouter_max_tokens = 2048
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key, base_url="https://openrouter.ai/api/v1")
```

```python
response = client.chat.completions.create(model=openairouter_model_name,messages=[{'role':'user', 'content':'List the prices for Epic Pass. format the output in json'}], max_tokens= 1000)
```
